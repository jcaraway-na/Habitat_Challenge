## Connecting to API

- api uri is stored
- request made to API, responce converted to json
- json file parsed to results and records. Then saved to a new dataFrame
- df.head() called to confirm data

```python
import pandas as pd
import requests as res


uri = 'https://data.nationalgrideso.com/api/3/action/datastore_search?resource_id=ddc4afde-d2bd-424d-891c-56ad49c13d1a'

data = res.get(uri).json()
df = pd.DataFrame(data['result']['records'])

df.head()
```

## Loading data to SQL

- PGAdmin call sqlalchemy and import engine
- engine created for postgres 
- dataFrame from above is loaded to newly created db table.

```python
from sqlalchemy import create_engine
db_password = 'Tacobell'
engine = create_engine(f"postgresql://postgres:{db_password}@127.0.0.1:5432/Hab")

df.to_sql('HabData',con=engine)
```

## Query SQL Table Select all records

- PG select all query to confirm records have been loaded to db table.

```python
select_all = engine.execute('SELECT * FROM "HabData"')
for r in select_all:
    print(r)
```
